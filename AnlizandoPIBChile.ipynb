```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL de la página web
url = "https://si3.bcentral.cl/indicadoressiete/secure/Serie.aspx?gcode=PIB_RE_EP03&param=NABaAHEASAB6AFoAeABlADYALQA3AHUAZQBTAE4ATgBSAHUAZwBqAEcASQBMAFQAeQBEACQAaQBpAFYAZAAxAEUAbgBZAFgAYQBhACQAcQBlAHkAYQBaADAARAAzAEsAbAAkAGIAWABqAEoANgAuAFQAUgBPAC4ANABZAFAAOQBZAFgAbABHAE0ARgBPADgAMgBBAGIAbwA0AC4AWgBHAEYARQA5AHEAOABwAHMAYwBCAG8AeQBOAG0AdABYAEoAdQBQAEgASwB6ADMAdgBMAFkASwA4AA=="
```

```python
# Realizar la solicitud GET a la página web y obtener el contenido HTML
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Encontrar la tabla con los datos
table = soup.find(id = 'gr')
```

```python
# Extraer los datos de la tabla y convertirlos en una lista de listas
data = []
for row in table.find_all("tr"):
    row_data = [cell.get_text(strip=True) for cell in row.find_all("td")]
    data.append(row_data)
print(data)
```

```python
# Crear un DataFrame con los datos
df = pd.DataFrame(data[1:], columns=data[4])
```

```python
# Imprimir el DataFrame
df
```

```python
# Suponiendo que ya tienes un DataFrame llamado df con tus datos

# Lista con los nombres de las columnas
nombres_columnas = ["AÑO", "1°", "2°", "3°", "4°"]

# Asignar los nombres de las columnas al DataFrame
df.columns = nombres_columnas

# Imprimir el DataFrame con los nombres de las columnas
print(df)
```

```python
# Crear un DataFrame con datos y nombres de columnas al mismo tiempo
df = pd.DataFrame(data[1:], columns=nombres_columnas)
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Obtener los años para el eje x
años = df['AÑO']
# Obtener los valores para cada trimestre
trimestre_1 = df['1°']
trimestre_2 = df['2°']
trimestre_3 = df['3°']
trimestre_4 = df['4°']
```

```python
# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))  # Ajusta el tamaño del gráfico según tus preferencias
# Graficar cada trimestre a lo largo de los años
plt.plot(años, trimestre_1, label='1° Trimestre')
plt.plot(años, trimestre_2, label='2° Trimestre')
plt.plot(años, trimestre_3, label='3° Trimestre')
plt.plot(años, trimestre_4, label='4° Trimestre')

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Miles de millones de pesos')
plt.title('Valores por Trimestre a lo largo de los Años')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
